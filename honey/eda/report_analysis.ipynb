## Recommendations 

- Lower all possible costs, such as shipping, transportation, storage to get APPO price close to 50 cents. This will allow Baron Honey Co. to compete with the top 5 best selling honey products on price and Baron Honey will have the advantage with taste.   
- Test our raw honey to understand where it stands on the medicinal rating scale compared to Manuka and other honey 
    - If our honey is highly rated we can then advertise it as such and increase the price per ounce 
- To overcome the high initial shipping costs a single serving product should be offered. 
    - This will increase margins and tape into a new market of long distance athletes, parents wanting to give their children healthy sweets, other athletes in general wanting a high carb source. 
- 16 oz jars should initially make up the majority of our product stock due to its popularity with customer experimentation. Once brand reputation has been established, the focus should shift to having the most stock of 32 oz jars. 
- Title and description of product listing should describe the honey as:
    - "Raw Unfiltered" 
    - "Sweetener" 
    - "Pure"
    - "Best tasting"
    - "Purity guarantee" 
    - "Natural"
